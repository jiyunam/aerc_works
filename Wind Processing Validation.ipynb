```python
# Set column names to be the same
data_1.columns = data_2.columns

# Set RPM type of data_2 to be the same 
data_2 = data_2.astype({'RPM': data_1['RPM'].dtypes})
```

```python
# Check if the two data
data_1.equals(data_2)
```
